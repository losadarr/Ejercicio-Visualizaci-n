```python
def factorial(n):
    
    if n < 0:
        return "número negativo"
    
    elif n == 1 or n == 0: #ejemplo mas sencillo
        return 1
    
    else:
        result=1
        for i in range(2,n+1):
            result*=i
        return result
numero=5
```

b) Crea una función que verifique si un número es primo o no. (0.6 puntos)

```python
def primo(numero):

    
    if numero < 2: #numero primo si mayor que 1
        return False
    
    for i in range(2, int(sqrt(numero)) + 1): #numero primo si mayor que 1, hasta su raiz mas 1
        
        if numero % i == 0:
            return False
        
    return True

numero=23
```

```python
if primo(numero):
    print("es un número primo")
else:
    print("no es un número primo")
```

c) Construye una Regresión lineal y un Random forest que predigan el Median house value según los datos disponibles. (0.75 puntos)

```python

# Definir características (X) y variable objetivo (y)
x = california_df.drop('Median House Value', axis=1)
y = california_df['Median House Value']

california_df




```

```python

#entrenamiento ybtest
x_train,x_test,y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#modelo regresion lineal
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
#predecir prueba reg lineal
y_pred_linear = linear_model.predict(x_test)

#modelo Random Forest
random_forest_model = RandomForestRegressor(random_state=42)
random_forest_model.fit(x_train, y_train)
#predecir prueba random forest
y_pred_rf = random_forest_model.predict(x_test)




```

```python
#plotly
#funcion evaluacion de modelo intern
def evaluacion(y_true, y_pred, modelo_nombre):
    
    
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{modelo_nombre} - MSE {mse:.2f}')
    print(f'{modelo_nombre} - R2 {r2:.2f}')
    
    

print("evaluacion de los modelos")

#visu reg lineal
fig_linear = px.scatter(x=y_test, y=y_pred_linear, labels={'x': 'valores reales', 'y': 'predicciones'},
                        title="REG LINEAL: valores reales vs predicciones")
fig_linear.show()

#visu random forest
fig_rf = px.scatter(x=y_test, y=y_pred_rf, labels={'x': 'valores reales', 'y': 'predicciones'},
                    title="RANDOM FOREST: valores reales vs predicciones")
fig_rf.show()
```
